```python
#Sum of square of even number in the list
result = sum(list(map(lambda n: n * n, list(filter(lambda n: n % 2 == 0, [1, 2, 3, 4, 5])))))
result
```

```python
#OOPS Concept

#Class and Objects

#Class is a blueprint

class car:
  engine = 1 # member variable
  tyre = 4
  door = 4
  seat = 4
  accelator = 1
  brake = 1

  # Define functions
  def apply_accelator(self):
    print("Accelator applied !!! Car is moving.")

  def apply_brake(self):
    print("Brake is applied!!! Car stopped.")
```

```python
#Exploratory Data Analysis(EDA) Technique -  explore/deep analysis all data
# EDA is a 3 type of analysis
#1. Uni variete analysis - Taking each column one by one and analysing data
    #Ex: Age is a numeric column - Min age, Max age, Average age, distribution of age.
    #Ex: Titanic - Fare - 890 people travlled. Fare ranges from $100-$800. Avg fare $100. Most people travlled in lowest class.
#2. Bi variate analyssis - Taking 2 columns and comparing data & checking relationship between 2 columns whether its a positive/negative relationship
    #Ex: As year passes, rent also increases. - Positive relationship
    #Ex: As year passes, car price decreases (depreciation) - Negative relationship
#3. Multi variate analysis - Finding relationship between more than 2 columns
    #Ex: Comparing bill, tips, time. During lunch time, how much billing happened and how much tip is payed.
    #Ex: During lunch time, how many male/female person is paying more tips
```

```python
import seaborn as sns
```

```python
# seaborn github dataset
df = sns.load_dataset('tips') # direcly load the dataset from github repo based on name
df
```

```python
#Relational plot
sns.relplot(data = df, x= 'total_bill', y = 'tip')
# x and y values should match column name in dataset
```

```python
sns.relplot(data = df, x= 'total_bill', y = 'tip', hue='sex')
# Hue - produce column values with different colors
```

```python
#Analysis from above: Male is paying more bill and tips
```

```python
sns.relplot(data = df, x= 'total_bill', y = 'tip', hue='sex', style='smoker')
# style - produce column values with different styles.
```

```python
sns.relplot(data = df, x= 'total_bill', y = 'tip', hue='sex', style='smoker', size='size')
#size - produce column values with different sizes.
```

```python
sns.relplot(data = df, x= 'total_bill', y = 'tip', hue='sex', style='smoker', size='size', row ='day')
# row - produce subsets of data based on column value
```

```python
sns.relplot(data = df, x= 'total_bill', y = 'tip', hue='sex', style='smoker', size='size', row ='day', col = 'time')
# col - produce subsets of data based on column value
```
