```python
#map(logic, list or series or multiple numbers)

#Goal to square every element in the list
sq = list(map(lambda n: n*n, [1,2,3,4,5]))
sq
```

```python
cube = list(map(lambda n: n*n*n, [1,2,3,4,5])) # Find cube of every number from the list
cube
```

```python
even = list(filter(lambda n: n%2==0, [1,2,3,4,5])) # Filter even number from the list
even
```

```python
odd = list(filter(lambda n: n%2!=0, [1,2,3,4,5])) # Filter odd number from the list
odd
```

```python
del sum # Delete previous variable sum used in user defined function
```

```python
#Sum of square of even number in the list
result = sum(list(map(lambda n: n * n, list(filter(lambda n: n % 2 == 0, [1, 2, 3, 4, 5])))))
result
```

```python
#Sum of square of odd numbers in the list
result = sum(list(map(lambda n: n * n, filter(lambda n: n % 2 != 0, [1, 2, 3, 4, 5]))))
result
```
